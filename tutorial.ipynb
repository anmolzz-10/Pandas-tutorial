### Loc and Iloc in Pandas
#### Syntax for loc: df.loc[name of row label, name of column labels]
#### Syntax for iloc: df.iloc[row index number, column index number]

```python
coffee.loc[1]
```

```python
coffee.loc[0:3] ##can also use slicing difference here is it is inclusive meaning the stop is also included unlike python list slicing
```

```python
coffee.loc[[1,4],['Day','Coffee Type']]
```

```python
coffee.loc[2:3, ['Day','Units Sold']]
```

```python
coffee.iloc[0:1, [0,2]] ##unlike loc its exclusive as it didnt include 1st index just 0th 
```

## Filtering Data

```python
bios.tail(5)
```

```python
bios.info()
```

```python
bios.loc[bios['height_cm'] >215,['name','height_cm']] ##using double square brackets to filter based on a given condition
```

```python
bios[bios['height_cm'] >215][['name','height_cm']] ##same can be done using df[condition] just add specific column names inside [['column_name']] afterwards to get specific columns
```

```python
bios.loc[(bios['born_country']== 'USA') & (bios['weight_kg'] >60), ['name','born_region']] ##adding parenthesis to add multiple conditions and required columns can be added
```

```python
bios.loc[(bios['height_cm']> 215) & (bios['born_country']== 'USA')]
```

```python
bios.loc[bios['name'].str.contains('Keith')] ##str.contains used to find a particular string inside a column
```

```python
bios.loc[bios['name'].str.contains('keith', case=  False)] ##irrespective of case
```

```python
bios.loc[bios['name'].str.contains('keith| patrick', case=  False)] ##| means or
```

```python
bios[bios['born_country'].isin(['USA','RUS','GBR'])]
```

```python
bios.loc[(bios['born_country'].isin(['USA','FRA','GBR'])) & (bios['name'].str.startswith('Keith'))] ##startswith used to filter starting words
```

## Adding/ Removing Columns

```python
coffee.head(5)
```

```python
coffee['price']= 4.99 ##add derived column based on given columnd
```

```python
coffee.head()
```

```python
import numpy as np
coffee['new_price']= np.where(coffee['Coffee Type']== 'Espresso', 3.99,5.99)
```

```python
coffee.head()
```

```python
coffee.drop(0) ##drop a particular row using index
```

```python
coffee.drop(columns= 'price') ##remove a column by specifying column
```

```python
coffee.drop(columns='price', inplace= True)  ##inplace is used for applying permanent change
coffee
##Alternate approach coffee= coffee.drop(columns= 'price') storing the cleaned data in same variable name
```

```python
coffee_new =coffee.copy()
coffee_new ##use copy() to actually copy the dataframe at the time of operation because once the data is re-executed, the copied data will lose out the results of operations unless all operations are repeated
```

```python
coffee ##like here for ex after coffee data was re-run it reverts back to its original form but coffee_new retained the operations due to copy() function
```

```python
coffee = coffee[['Day','Coffee Type','Units Sold', 'new_price']] ##remove columns by only accessing required columns and storing inside variable of same name
```

```python
coffee
```

```python
coffee['revenue']= coffee['Units Sold'] * coffee['new_price']
```
