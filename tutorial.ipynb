### Important Use Case

```python
coffee = pd.DataFrame({
    'Shop': ['BrewCo', 'Daily Grind', 'BrewCo', 'Cafe Hero', 'Cafe Hero'],
    'Coffee Type': ['Espresso', 'Latte', 'Latte', 'Espresso', 'Cappuccino'],
    'Units Sold': [100, 120, 130, 90, 110],
    'Price': [3.99, 4.99, 4.99, 3.99, 4.49]
})

coffee['Revenue'] = coffee['Units Sold'] * coffee['Price']
coffee
```

#### If its asked to calculate average of "TOTAL" revenue of a shop 

```python
##Normal mind says just use group by and mean

coffee.groupby('Shop')['Revenue'].mean().reset_index()
```

##### But this is incorrect because its actually average as per each transaction or row and not Total revenue.
##### For ex: If we use just mean and group by shop it will give average of each shop separately and not the average of total revenue.



## Correct way



```python
##Correct way

print(coffee.groupby('Shop')['Revenue'].sum().mean())

##Or Step by step
grouped= coffee.groupby('Shop')['Revenue'].sum() ##adds up all revenue as per shop 
avg_shop = grouped.mean()
print(avg_shop)
```

#### And just taking the mean of Revenue column is also wrong because a thought can come if we r just adding up all revenue why not just take revenue mean

#### Size and Unstack

##### This groups the data by two columns:

first by release_year

then by type

So it creates groups like:

(2020, Movie)

(2020, TV Show)

(2021, Movie)

(2022, TV Show)

```python
df.groupby(['release_year', 'type'])
'''
release_year  type    
2020          Movie      
              TV Show
    
2021          Movie      
2022          TV Show   
'''
```

##### ✅ 2. .size()
This counts the number of rows in each group.

So we know how many movies/TV shows were released per year.

```python
df.groupby(['release_year', 'type']).size()

'''
release_year  type    
2020          Movie      1
              TV Show    1
2021          Movie      2
2022          TV Show    1
dtype: int64'''
```

#### 3. .unstack()
Now we reshape the Series into a table, where:

Rows = release_year

Columns = type (unstacked)

Cells = count of rows

```python
df.groupby(['release_year', 'type']).size().unstack()

'''
type           Movie  TV Show

release_year                  
2020              1        1
2021              2      NaN
2022            NaN        1
'''
```

#### 4. .fillna(0)
Replaces NaN (missing combinations) with 0.

```python
'''
type           Movie  TV Show
release_year                  
2020              1        1
2021              2        0
2022              0        1
'''
```

#### Alternative use pivot tables

```python
import pandas as pd

df = pd.DataFrame({
    'Date': ['2023-01', '2023-01', '2023-02', '2023-02', '2023-02'],
    'Category': ['A', 'B', 'A', 'B', 'B'],
    'Sales': [100, 200, 150, 250, 180]
})
```

```python
df.groupby(['Date','Category'])['Sales'].sum().reset_index()
```

```python
df.pivot(index='Date', columns='Category', values='Sales')
```

```python
df.pivot_table(index='Date', columns='Category', values='Sales', aggfunc='sum')

##Intuition:

##For each (Date, Category) combo, it sums the Sales

##If multiple rows exist → it aggregates them

```

## Analogy:
Think of pivot() as:

“I want a table with rows and columns and I already know there’s only one value per cell.”

And pivot_table() as:

“I want a table, but I don’t mind if there are duplicates — just summarize them.”
